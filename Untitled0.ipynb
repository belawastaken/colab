Wir analysieren die Apple Aktie im Jahr 2025. Dabei betrachten wir Kursverläufe, tägliche Renditen, Volatilitäten und die Korrelation zu Wettbewerbern.

```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import seaborn as sns
import numpy as np
```



In diesem Teil wurden die Aktiendaten für Apple für den Zeitraum von 01.01.2025 bis zum 31.12.2025 von yfinance herruntergeladen und wiedergegeben. Zusätlich wurden die Daten der schließtage direkt fallen gelassen um Lücken in Folgenden Darstellungen zu vermeiden. Im Folgenden werden wir mit diesen Daten weiterarbeiten.

Die Tabelle stellt die täglichen Schwankungen der Apple Aktie über das Jahr 2025 dar.
Die Aktie verhält sich meist stabil, zeigt aber im Frühjahr eine außergewöhnlich heftige Kursbewegung nach unten und oben.

```python
Data_Apple = yf.download(
    tickers=["AAPL"],
    start="2025-01-01",
    end="2025-12-31").dropna()
print(Data_Apple)
```

In Folgender Darstellung, sieht man den Kursverlauf von Apple. Dazu haben wir die heruntergeladenen Daten auf den Zeitpunkt der Schließung der Banken gesetzt um eine repräsentative Datenerhebung zu erreichen.

Die Grafik zeigt eine Volatilitätsanalyse der Apple Aktie für das Jahr 2025.
Die Renditen sind weitgehend normalverteilt um den Nullpunkt, weisen jedoch vereinzelt extreme Ausreiser auf.



Erklärung der Hebungen und Senkungen:

```python
Data_Apple_cl=Data_Apple["Close"]
Data_Apple_cl_pct = Data_Apple_cl.pct_change()
Data_Apple_cl_pct.plot()
plt.title("Kursverlauf")
```

Die Streuung beträgt ca. 6,5 % des Mittelwerts

```python
Standardabweichung=Data_Apple_cl_pct.std()
Standardabweichung *np.sqrt(252)
```

```python
Data_Applelong=yf.download(
    tickers=["AAPL"],
    start="2020-01-01",
    end="2025-12-31")
Data_Applelong_cl=Data_Applelong["Close"]
Data_Applelong_clpct=Data_Applelong_cl.pct_change()
Standardabweichung_rolling_year=Data_Applelong_clpct.rolling(window=252).std()*np.sqrt(252)
Standardabweichung_rolling_20Days=Data_Applelong_clpct.rolling(window=20).std()*np.sqrt(252)
```

Das obere Digramm zeigt die täglichen Renditen für das Jahr 2025.
Ein extremer Ausschlag im April deutet auf ein kurzes, heftiges Marktereignis hin.

Das untere Diagramm vergleicht die kurz- und langfristige Volatilität seit 2020.

```python
fig,(graph1,graph2)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))
graph1.plot(Data_Apple_cl_pct,color="brown")
graph1.set_title("Tägliche Renditen")
graph1.grid(alpha=0.5)
graph2.plot(Standardabweichung_rolling_20Days,label="Volatilität 20 Tage",color="brown")
graph2.plot(Standardabweichung_rolling_year,label="Volatilität Jahr",color="orange")
graph2.set_title("Volatilitäten Vergleich")
graph2.set_xlabel("Datum")
graph2.set_ylabel("Volarität")
graph2.legend()
graph2.grid(alpha=0.5)
```

```python
Data_alle=yf.download(
    tickers=["AAPL","XIACY","SONY","005930.KS"],
    start="2025-01-01",
    end="2025-12-31",
    auto_adjust=False)
Data_alles= Data_alle.dropna()
```

```python
ticker_names={"AAPL":"Apple",
              "XIACY":"Xiaomi",
              "SONY":"Sony",
              "005930.KS":"Samsung"}
```

Wir betrachten den Kursverlauf von Samsung, Apple, Sony und Xiaomi, um zu sehen, wie sich Apple im Vergleich zu anderen großen Unternehmen entwickelt hat.
Alle wurden zum Start auf den gleichen Wert gesetzt.
Dabei zeigt sich Samsung als klarer Spitzenreiter mit mehr als einer Verdoppelung des Werts, während Apple und Sony das Jahr fast ohne Gewinn beenden.

```python
Data_alle_cl = Data_alles["Close"]
Data_alle_cl_pct = Data_alle_cl.pct_change(fill_method=None)
rebas_data = (Data_alle_cl / Data_alle_cl.iloc[0]) * 100
def label(t):
    return f"{ticker_names.get(t, t)} ({t})"
for column in rebas_data.columns:
    plt.plot(rebas_data[column], label=label(column))
plt.xlabel("Datum")
plt.title("Konkurrenzverhalten")
plt.ylabel("Aktienhöhe")
plt.grid(alpha=0.5)
plt.legend()
```

Diese Heatmap zeigt die Korrelationen zwischen Apple, Samsung, Xiaomi und Sony. Die Korrelation von 1 ist in einer Heatmap unvermeidbar, da sie die Korrelation von der Marke zu sich selbst anzeigt.

Die Heatmap visualisiert die Korrelationen zwischen den Aktienkursen von Samsung, Apple, Sony und Xiaomi.
Samsung und Apple weisen mit einem Wert von 0,84 die stärkste positive Abhängigkeit auf.
Im Gegensatz dazu zeigt Xiaomi eine negative Korrelation zu Samsung und Apple. Das deutet darauf hin, dass sich dieser Aktienkurs tendenziellentgegengesetzt zu den beiden Marktführern bewegt.

```python
sns.heatmap(Data_alle_cl.corr(),annot=True,cmap="copper")
plt.title("Korrelation")
```

Dieses Histogramm zeigt die Korrelation von Apple zu Samsung in einem Histogramm um die Korrelation besser darstellen zu können. Wie haben diese Konkurrenten gewählt, weil sie die höhste Korrelation haben.

Das 2D-Histogramm veranschaulicht die Verteilung der täglichen Renditen von Apple und Samsung.
Es zeigt, dass sich die Aktien von Apple und Samsung meistens sehr ähnlich verhalten, da sich die meisten hellen Punkte in der Mitte treffen.
Weil die Punktewolke eher hoch als breit ist, erkennt man, dass der Kurs von Samsung deutlich stärker rauf und runter springt als der von Apple.

```python
df_for_hist = Data_alle_cl_pct[['AAPL', '005930.KS']].dropna()
plt.hist2d(df_for_hist['AAPL'], df_for_hist['005930.KS'], bins=50, cmap='Oranges_r')
plt.xlabel('Tägliche Rendite AAPL')
plt.ylabel('Tägliche Rendite 005930.KS')
plt.title('2D Histogramm der täglichen Renditen von AAPL und Samsung')
plt.colorbar(label='Häufigkeit')
plt.show()
```


Diese Histogramm zeigt die negative Korrelation von Apple zu Xiaomi.

Das 2D Histogramm visualisiert die Häufigkeitskeitsverteilung der täglichen Renditen von Apple und Xiaomi. Die hellen Bereiche signalisieren eine hohe Konzentration von Datenpunkten um den Nullpunkt.

```python
df_for_hist = Data_alle_cl_pct[['AAPL', 'XIACY']].dropna()
plt.hist2d(df_for_hist['AAPL'], df_for_hist['XIACY'], bins=50, cmap='Oranges_r')
plt.xlabel('Tägliche Rendite AAPL')
plt.ylabel('Tägliche Rendite Xiaomi')
plt.title('2D Histogramm der täglichen Renditen von AAPL und Xiaomi')
plt.colorbar(label='Häufigkeit')
plt.show()
```
